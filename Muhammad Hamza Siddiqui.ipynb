```python
import pandas as pd #library for reading our data
import matplotlib.pyplot as plt #library for drawing our figures
def population(P):#function to display population trend
      x=P['Year'] #on x-axis is the year
      y=P['Population'] #on y-axis is the population
      plt.bar(x,y,color="Blue") #figure is plotted with blue bars
      plt.title('United Kingdom Population') #this is title of our chart
      plt.xlabel('Year') #this is x label
      plt.ylabel('Population') #this is y label in our figure
      plt.show()

def mobile(M):
      x=M['Year']#on x-axis is the year
      y=M['Value']#on y-axis is the value
      M.plot.box(grid='True')
      #plt.boxplot(M)#figure is plotted with green bars
      plt.title('United Kingdom Mobile users')#this is title of our chart
      plt.xlabel('Year')#this is x label
      plt.ylabel('Value')#this is y label
      plt.show()

def bss(B):
      x=B['Year']#on x-axis is the year
      y=B['Value']#on y-axis is the value
      plt.scatter(x,y,color="Purple")#figure is plotted with purple bars
      plt.title('New business registered in United Kingdom')#this is title of our chart
      plt.xlabel('Year')#this is x label
      plt.ylabel('Value')#this is y label
      plt.show()

s1=pd.read_excel('datascience1.xlsx','Sheet1')#reading population data
s2=pd.read_excel('datascience1.xlsx','Sheet2')#reading mobile users data
s3=pd.read_excel('datascience1.xlsx','Sheet3')#reading new business data
population(s1)#function called to draw population trend
mobile(s2)#function called to draw mobile users trend
bss(s3)#function called to draw new business registered trend
```
